```python
'''the raw file nyse_stocks.csv was downloaded from the internet here:
https://old.nasdaq.com/screening/companies-by-name.aspx?letter=0&exchange=nasdaq&render=download

we will clean this file to create a list of stock tickers to input into the spider's url list
'''

f = open('nyse.csv', 'r')
data = f.readlines()
f.close()
```

```python
data = list(map(lambda x: str.split(x),data))
data
```

```python
#experimenting with the data input to clean it accordingly
data[1][0].split(',')[0].replace('"','')
```

```python
#create new list lis to capture only the stock ticker from each row of data
lis=[]
for l in range(len(data)):
    lis.append(data[l][0].split(',')[0].replace('"',''))
    
#get rid of the header row
lis = lis[1:]
lis
```

```python
#write a new file to save the stock ticker list, seperate each ticker with ','
f = open('nyse_cleaned.csv','w')
for line in lis:
    f.write(line+',')
f.close()
```
